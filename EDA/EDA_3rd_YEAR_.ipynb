**METHOD 1 MEAN OR MEDIAN-MISSING VALUE**

---



```python
d1= df.copy()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
for i in data_keys:
  X = np.array(d1[i]).reshape(-1,1)
  imp.fit(X)
  X_prep = imp.transform(X)
  d1[i] = X_prep.reshape(1,-1)[0]

d1.head()  
```

```python
total_cells= np.product(d1.shape)
total_missing=d1.isnull().sum().sum()
# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
d1.describe()
```

**METHOD 2 Interpolation**

```python
d2=df.copy()
for i in data_keys:
  d2[i].interpolate(method='linear', direction = 'forward', inplace=True) 

d2.head()
```

```python
total_cells= np.product(d2.shape)
total_missing=d2.isnull().sum().sum()
# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

# OverSamppling :SMOTE

---



```python
from imblearn.over_sampling import SMOTE
```

```python
X=d1.drop(labels=['class'], axis=1)
y=d1['class']
```

```python
oversample = SMOTE()
X,y=oversample.fit_resample(X,y)
```

```python
sns.countplot(x=y)
```

```python
data_x= pd.DataFrame(X,columns=data_keys[:len(data_keys)-1])
print(data_x.head())
#y class data 
data_x.shape
```
